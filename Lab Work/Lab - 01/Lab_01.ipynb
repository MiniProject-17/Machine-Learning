```python
df = pd.read_csv("mumbai-monthly-rains.csv")
# To display the top  rows
df.head()
```

```python
df.dtypes
```

```python
df = df.drop(['Jan', 'May','June','July','Aug','Sept','Oct','Nov','Dec'], axis=1)
df.head(5)
```

```python
df = df.rename(columns={"Feb": "February"})
df.head(5)
```

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df.count()
```

```python
df = df.drop_duplicates()
df.head(5)
```

```python
df = df.drop_duplicates()
df.head(5)
```

```python
df.count()
```

```python
## Dropping missing or null values
print(df.isnull().sum())
```

```python
df = df.dropna()    # Dropping the missing values.
df.count()
```

```python
print(df.isnull().sum())   # After dropping the values
```

```python
## Detecting outliers
sns.boxplot(x=df['April'])
```

```python
sns.boxplot(x=df['Mar'])
```

```python
sns.boxplot(x=df['February'])
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

```python
# Create the histogram
df['April'].plot(kind='hist', figsize=(10, 5))
plt.title("Rainfall in April")
plt.ylabel('Frequency')  # Or 'Count'
plt.xlabel('Rainfall (mm)')  # Adjust units if needed

plt.show()  # Display the histogram
```

```python
df['Mar'].plot(kind='hist', figsize=(10, 5))
plt.title("Rainfall in April")
plt.ylabel('Frequency')  # Or 'Count'
plt.xlabel('Rainfall (mm)')  # Adjust units if needed
```

```python
df['February'].plot(kind='hist', figsize=(10, 5))
plt.title("Rainfall in April")
plt.ylabel('Frequency')  # Or 'Count'
plt.xlabel('Rainfall (mm)')  # Adjust units if needed
```
